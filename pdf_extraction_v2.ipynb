# Potential Section Extraction

```python
from PyPDF2 import PdfReader, PdfWriter
import pandas as pd
import csv
import os
import fitz
from re import search
```

```python
def page_finder(protocol):
    # potential keywords in sections
    potential_sections = [
    "objective",
    "study endpoints",
    "event",
    "assessments",
    "activities",
    "abbreviations",
    "endpoint",
    "evaluation",
    "measure",
    "design",
    "synopsis",
    "questionnaire",
    "outcome",
    "patient reported",
    "definitions",
    "flow chart",
    "visits",
    "schedule",
    ]
    # table of contents
    # input_pdf = PdfReader(open(protocol, 'rb'))
    try:
        pdf = fitz.open(protocol)
        toc = pdf.get_toc()
        # header = page number
        section_titles = {}
        for item in toc:
            section_titles[item[1]] = item[2]
        # page numbers to extract
        page_nos = []
        for title in section_titles.keys():
            for potential in potential_sections:
                if search(potential, title.lower()):
                    page_nos.append(section_titles[title])
        page_nos = list(dict.fromkeys(page_nos))
        page_nos2 = page_nos.copy()
        for page in page_nos:
            page_nos2.append(page + 1)
        page_nos2 = list(dict.fromkeys(page_nos2))
        # try:
        #     if page_nos2[-1] >= len(input_pdf.pages):
        #         page_nos2.remove(page_nos2[-1])
        # except:
        #     print(protocol, "finder")
        return page_nos2
    except:
        print(protocol)
```

```python
def csv_writer(page_nos, inpath, outpath):
    df = pd.read_csv(inpath, encoding='utf-8', on_bad_lines='skip', lineterminator="\n")
    df = df.astype(str)
    df.set_index('Unnamed: 0')
    not_page_numbers = []
    for row in df.index:
        if row not in page_nos:
            not_page_numbers.append(row)
    df = df.drop(df.index[not_page_numbers])
    with open(outpath, 'wb') as f:
        df.to_csv(f, index=False)
    print("ok")
```

```python
def pdf_writer(inpath,  outpath, page_nos):
    input_pdf = PdfReader(open(inpath,'rb'))
    output_pdf = PdfWriter()
    for i in page_nos:
        page = input_pdf.pages[i]
        output_pdf.add_page(page)
    with open(outpath, 'wb') as f:
        output_pdf.write(f)
        ok_count += 1
        print("ok")
```
